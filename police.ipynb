# The subplot() Function
The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

# Creating Bars

A bar chart (or bar graph) is a type of data visualization that uses rectangular bars to represent data values. 
Each bar's length or height is proportional to the value it represents, 
allowing for easy comparison between categories.

# Key Features of a Bar Chart:
Bars: Represent individual categories (e.g., countries) and their values (e.g., population).

Axes:
The horizontal axis (x-axis) typically represents categories (e.g., country names).

The vertical axis (y-axis) represents numerical values (e.g., population size).

Bar Height/Length: Corresponds to the value of the category (the taller or longer the bar, the higher the value).

Explanation 
bars: Holds the bar objects created by plt.bar().

plt.text(): Adds text labels on top of each bar:

bar.get_x() + bar.get_width()/2. centers the label on the bar.

height * 1.01 positions the label slightly above the bar.

f'{height:,}' formats the population with commas for better readability.

ha='center' and va='bottom': Align the text horizontally in the center and vertically at the bottom of the text.

```python
import matplotlib.pyplot as plt
# sample data: country names and thier populations
countries = ['china', 'india', 'USA', 'indonesia', 'pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]
plt.figure(figsize=(10,6))
bars = plt.bar(countries,population, color = 'pink')

# Adding title and labels
plt.title('population of countries', fontsize=16)
plt.xlabel('countries')
plt.ylabel('population')

# Adding labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2., height*1.01, #Adjust position
             f'{height:,}', ha='center', va='bottom', fontsize=10)
```

# Horizontal bar chart

```python
# country names and thier population
import matplotlib.pyplot as plt
countries = ['china', 'india', 'USA', 'indonesia', 'pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]

# create a horizontal bar chart
plt.figure(figsize=(10,6))
bars = plt.barh(countries,population, color = 'green')

# Adding title and labels
plt.title('population of countries', fontsize=16)
plt.xlabel('population', fontsize =12)
plt.ylabel('countries', fontsize=12)


# Adding labels on the right side of each bar
for bar in bars:
    width = bar.get_width()
    height = bar.get_height()
    plt.text(width * 1.01,
             bar.get_y() + height /2., #Adjust position
             f'{width:,}', ha='left', va='center', fontsize=10)
```
